```python
%matplotlib inline
import numpy as np
import matplotlib as plt
import pandas as pd
import re
from datetime import datetime
from dateutil import parser
```

```python
def findOid(str):
    pars = re.match('([\w*\s*]*)(Oid)\:\s*([\d*\.]*)\s*(Value)\:\s*([\d*\.]*)', str)
    if pars:
        return [pars[1], pars[2], pars[3], pars[4], pars[5]]
    return [str]

def process(pars):
    p1 = parser.parse(pars[1])
    p3 = pars[3].replace('[','').replace(']','')
    p4 = pars[4].replace('*','').replace('  ',' ')
    arr = [p1, pars[2], p3]
    pf = findOid(p4)
    arr.extend(pf)
    return arr
```

```python
py_array = []
i=0
extra_cols = 0
row = []
with open('./Datasets/fault_public.log') as f:
    for line in f:
        pars = re.match('(\d{4}-\d{2}-\d{2}\s\d{2}:\d{2}:\d{2},\d{3})\s(\w*)\s*(\[.*\])\s(.*)', line)
        if pars is None:
            row.extend(line.strip().split(':', 1))
        else:
            if row:
                py_array.append(row)
            row = process(pars)
```

```python
max_cols = 0
for line in py_array:
    if max_cols < len(line):
        max_cols = len(line)

cols = ['P%d' % (i+1) for i in range(max_cols)]

df = pd.DataFrame(py_array, columns=cols, dtype=float)
df
```
